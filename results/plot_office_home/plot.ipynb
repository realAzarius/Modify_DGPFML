```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_path = [
    'pretrain_resnet_office-home_FedAvg_64_5_0.01_0.01_5c_1000r.h5',
    'office-home_PerAvg_conv4_64_5_0.01_0.01_5c_1000r.h5',
    # 'office-home_PerAvg_16_5_0.01_0.01_4c_500r.h5',
    'office-home_pFedInit_64_5_0.01_0.005_5c_600r.h5'
]
# label = ['FedAvg', 'Per-FedAvg(0.01)', 'Per-FedAvg(0.005)', 'pFedInit']
label = ['FedAvg', 'Per-FedAvg','pFedInit']

colors = ['#1B1A17', '#0F4C75', '#E45826', '#E6D5B8']

markers = ['x', 'o', 'P']
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))


max_testacc = []

for path, lab, col, mk in zip(data_path, label, colors, markers):
    data = h5py.File(path, 'r')
    losses = data['rs_train_loss'][::]
    train_acc = data['rs_train_acc'][::]
    test_acc = data['rs_glob_acc'][::]

    # print(np.mean(train_acc-test_acc))
    max_testacc.append(np.max(test_acc))
    
    x = [i*100 for i in range(len(losses+1))]

    ax1.plot(x, test_acc, label = lab, color=col, marker=mk, markevery=1, markersize=8)


ax1.set_xlabel("rounds")
ax1.set_ylabel("TEST_acc")
ax1.set_title('test')

# ax2.set_xlim(0.1, 0.7)
ax2.barh(range(len(data_path)), max_testacc, height=0.6, color='#0F4C75', zorder=10)
ax2.set_yticks(range(len(data_path)), ['FedAvg','Per-FedAvg', 'pFedInit'])
ax2.set_xlabel("TEST")
ax2.set_title("DIFF NON-IID")

for x, y in enumerate(max_testacc):
    ax2.text(y + 0.01, x-0.01, '%s' % (np.round(y, decimals=3)))

ax1.grid()
ax1.legend()
ax2.grid(zorder=0)
plt.savefig('femnist_acc_few_comp.pdf', dpi=600, pad_inches = 0)
```
